### 4.1.4.1 คลาสไดอะแกรม (Class Diagram) 

**Class diagram** เป็นไดอะแกรมในภาษา UML (Unified Modeling Language ) ที่ใช้อธิบายโครงสร้างของระบบโดยแสดงให้เห็นคลาส (Class) ต่างๆ ในระบบ 

คลาสไดอะแกรมเป็นเครื่องมือสำคัญในการออกแบบเชิงวัตถุ (Object-Oriented Design) ซึ่งภายในคลาสไดอะแกรมประกอบด้วย 

1. คลาส ซึ่งอธิบายด้วยรูปสี่เหลี่ยมที่ประกอบด้วยส่วนต่าง ๆ ดังต่อไปนี้

- **ส่วนบน**   ชื่อคลาส (Class Name)
- **ส่วนกลาง**  คุณสมบัติ (Attributes) เช่น ชื่อ, อายุ, ราคา
- **ส่วนล่าง**  การกระทำ (Operations/Methods) เช่น เดิน, กิน, ซื้อ 

ส่วนล่างของคลาสสามารถจำแนกได้ 2 อย่าง คือ
- **คุณสมบัติ** (Attributes)| คุณลักษณะของคลาส เช่น ชื่อ, อายุ, สี, ราคา แสดงอยู่ในส่วนกลางของสี่เหลี่ยมคลาส
- **การกระทำ** (Operations/Methods)| การกระทำหรือฟังก์ชันที่คลาสสามารถทำได้ เช่น เดิน, กิน, คำนวณ, ซื้อ แสดงอยู่ในส่วนล่างของสี่เหลี่ยมคลาส


2. ความสัมพันธ์ระหว่างคลาส ซึ่งอธิบายด้วยการใช้เส้นและลูกศรเชื่อมโยงระหว่างคลาส ที่ทำให้เกิดเป็นระบบขึ้นมา ความสัมพันธ์ระหว่างคลาสต่าง ๆ มีหลายประเภท เช่น 

- **Association**   ความสัมพันธ์ทั่วไประหว่างคลาส เช่น ลูกค้ามีความสัมพันธ์กับใบสั่งซื้อ (ลูกค้าสั่งซื้อสินค้า)
- **Inheritance/Generalization**  คลาสหนึ่งสืบทอดคุณสมบัติและการกระทำจากอีกคลาสหนึ่ง เช่น รถยนต์สืบทอดคุณสมบัติจากยานพาหนะ (รถยนต์เป็นยานพาหนะชนิดหนึ่ง)
- **Aggregation** คลาสหนึ่งเป็นส่วนประกอบของอีกคลาสหนึ่ง แต่สามารถมีอยู่ได้โดยอิสระ เช่น รถยนต์ประกอบด้วยเครื่องยนต์ (เครื่องยนต์สามารถอยู่ได้โดยไม่มีรถยนต์ เช่นไปอยู่ในเรือหางยาว)
- **Composition**  คลาสหนึ่งเป็นส่วนประกอบของอีกคลาสหนึ่งและไม่สามารถมีอยู่ได้โดยอิสระ เช่น  คณะประกอบด้วยภาควิชาต่าง ๆ (ภาควิชาไม่สามารถยังคงอยู่ได้ถ้าไม่มีคณะ)
- **Dependency**  คลาสหนึ่งใช้อีกคลาสหนึ่งในการทำงาน เช่น คลาสการชำระเงินพึ่งพาคลาสข้อมูลลูกค้า

**การใช้งานคลาสไดอะแกรม**

- สื่อสารและทำความเข้าใจระบบ: ช่วยให้ทีมพัฒนาและผู้เกี่ยวข้องเข้าใจโครงสร้างของระบบร่วมกัน
- ออกแบบระบบ: ใช้เป็นแบบแผนในการพัฒนาระบบ
- วิเคราะห์และแก้ไขปัญหา: ช่วยในการวิเคราะห์ปัญหาและหาแนวทางแก้ไข
- สร้างเอกสารประกอบ: ใช้เป็นเอกสารประกอบระบบ

#### 4.1.4.2 Sequence Diagram

**Sequence Diagram ** หรือ แผนภาพลำดับ  เป็นไดอะแกรมใน Unified Modeling Language (UML) ที่ใช้อธิบายลำดับการทำงานหรือปฏิสัมพันธ์ระหว่างอ็อบเจ็กต์ (Object) ต่างๆ ในระบบ โดยเน้นที่ลำดับเวลาของการส่งข้อความ (Message) หรือการเรียกใช้ฟังก์ชันต่างๆ ไดอะแกรมนี้ช่วยให้เห็นภาพรวมของการทำงานของระบบในแต่ละกรณีการใช้งาน (Use Case) ได้อย่างชัดเจน

องค์ประกอบหลักของ Sequence Diagram:

- **Object**   แสดงด้วยสี่เหลี่ยมผืนผ้า มีชื่ออ็อบเจ็กต์อยู่ภายใน เช่น :Customer, :Order, :Product
- **Lifeline**   เส้นแนวตั้งประที่อยู่ใต้กล่องอ็อบเจ็กต์ แสดงช่วงเวลาที่อ็อบเจ็กต์นั้นมีอยู่และทำงาน
- **Activation Bar/Focus of Control**   สี่เหลี่ยมผืนผ้าแคบๆ ที่วางอยู่บนเส้นชีวิต แสดงช่วงเวลาที่อ็อบเจ็กต์กำลังทำงานหรือประมวลผล
- **Message**    ลูกศรที่เชื่อมระหว่างเส้นชีวิตของอ็อบเจ็กต์ แสดงการส่งข้อความหรือการเรียกใช้ฟังก์ชัน มีหลายประเภท เช่น
    - *Synchronous Message*  มีลักษณะเป็นลูกศรทึบ มีหัวลูกศรแบบทึบ แสดงการเรียกใช้ฟังก์ชันที่ต้องรอผลลัพธ์ก่อนจึงจะทำงานต่อไปได้
    - *Asynchronous Message*   มีลักษณะเป็นลูกศรแบบเส้นประ มีหัวลูกศรแบบโปร่ง แสดงการเรียกใช้ฟังก์ชันที่ไม่ต้องรอผลลัพธ์ ทำงานต่อไปได้ทันที
    - *Return Message*  มีลักษณะเป็นลูกศรแบบเส้นประ มีหัวลูกศรแบบโปร่ง กลับจากแถบโฟกัสการควบคุม แสดงผลลัพธ์ของการเรียกใช้ฟังก์ชัน
- **Actor** แสดงด้วยรูปคน เป็นผู้ที่เริ่มต้นการทำงานหรือมีปฏิสัมพันธ์กับระบบ

**การใช้งาน Sequence Diagram**

- อธิบายลำดับการทำงานของระบบในแต่ละ Use Case
- วิเคราะห์และออกแบบปฏิสัมพันธ์ระหว่างอ็อบเจ็กต์
- สื่อสารและทำความเข้าใจการทำงานของระบบร่วมกัน
- สร้างเอกสารประกอบระบบ

#### 4.1.4.3 Usecase Diagram

**Use Case Diagram** หรือ แผนภาพกรณีการใช้งาน เป็นไดอะแกรมใน Unified Modeling Language (UML) ที่ใช้อธิบายการทำงานของระบบจากมุมมองของผู้ใช้งาน (Actor) โดยแสดงให้เห็นว่าผู้ใช้งานสามารถทำอะไรกับระบบได้บ้าง และระบบตอบสนองต่อการกระทำเหล่านั้นอย่างไร ไดอะแกรมนี้เน้นที่ฟังก์ชันการทำงานของระบบ ไม่ได้เน้นที่การทำงานภายในของระบบ

**องค์ประกอบหลักของ Use Case Diagram**

- **Actor** แสดงด้วยรูปคน เป็นผู้ที่ใช้งานระบบหรือมีปฏิสัมพันธ์กับระบบ อาจเป็นคน, ระบบอื่น หรืออุปกรณ์ภายนอก เช่น ลูกค้า, พนักงานขาย, ระบบธนาคาร
- **Use Case** แสดงด้วยวงรี เป็นฟังก์ชันหรือการกระทำที่ผู้ใช้งานสามารถทำกับระบบ เพื่อให้บรรลุเป้าหมายบางอย่าง เช่น สั่งซื้อสินค้า, ลงทะเบียน, ชำระเงิน
- **Relationship**  แสดงความสัมพันธ์ระหว่าง Actor กับ Use Case และระหว่าง Use Case ด้วยกันเอง มีหลายประเภท เช่น
    - *Association*  เขียนด้วยเส้นตรงที่เชื่อมระหว่าง Actor กับ Use Case แสดงว่า Actor นั้นสามารถใช้งาน Use Case นั้นได้
    - *Include* เขียนด้วยเส้นประพร้อมหัวลูกศรชี้ไปยัง Use Case ที่ถูกรวม แสดงว่า Use Case หนึ่งเรียกใช้การทำงานของอีก Use Case หนึ่งเสมอ เช่น การ "เข้าสู่ระบบ" อาจถูกรวมอยู่ใน Use Case "สั่งซื้อสินค้า"
    - *Extend* เขียนด้วยเส้นประพร้อมหัวลูกศรชี้จาก Use Case ที่ขยายไปยัง Use Case ต้นแบบ แสดงว่า Use Case หนึ่งขยายการทำงานของอีก Use Case หนึ่งภายใต้เงื่อนไขบางอย่าง เช่น "ชำระเงินด้วยบัตรเครดิต" อาจขยาย Use Case "ชำระเงิน"
    - *Generalization* เขียนด้วยเส้นทึบพร้อมหัวลูกศรสามเหลี่ยม แสดงความสัมพันธ์แบบสืบทอด เช่น Actor "ลูกค้าทั่วไป" และ Actor "สมาชิกระดับ VIP" โดย "สมาชิกระดับ VIP" สืบทอดคุณสมบัติจาก "ลูกค้าทั่วไป"

**การใช้งาน Use Case Diagram**

- กำหนดขอบเขตของระบบ (System Boundary)
- ระบุฟังก์ชันการทำงานของระบบจากมุมมองของผู้ใช้งาน
- สื่อสารความต้องการของระบบกับผู้เกี่ยวข้อง
- เป็นพื้นฐานในการสร้างไดอะแกรมอื่นๆ เช่น Sequence Diagram

#### 4.1.4.4  State Diagram

**State Diagram** หรือ แผนภาพสถานะ ใช้อธิบายพฤติกรรมของอ็อบเจ็กต์ (Object) โดยแสดงสถานะ (State) ต่างๆ ที่อ็อบเจ็กต์สามารถมีได้ รวมถึงการเปลี่ยนแปลงสถานะ (Transition) ที่เกิดขึ้นเมื่อมีเหตุการณ์ (Event) หรือเงื่อนไข (Condition) มากระตุ้น 

ไดอะแกรมนี้มีประโยชน์ในการทำความเข้าใจและออกแบบระบบที่มีพฤติกรรมแบบไดนามิกหรือมีการเปลี่ยนแปลงสถานะตามเวลา

__องค์ประกอบหลักของ State Diagram__

- **State**  แสดงด้วยสี่เหลี่ยมมุมมน แทนสถานะที่อ็อบเจ็กต์สามารถอยู่ได้ในขณะใดขณะหนึ่ง เช่น "เปิด", "ปิด", "รอ", "กำลังประมวลผล"
- **Initial State** แสดงด้วยวงกลมทึบ เป็นสถานะแรกที่อ็อบเจ็กต์เริ่มต้น
- **Final State** แสดงด้วยวงกลมซ้อนวงกลม เป็นสถานะสุดท้ายที่อ็อบเจ็กต์สิ้นสุดการทำงาน
- **Transition** แสดงด้วยลูกศรที่เชื่อมระหว่างสถานะต่างๆ แทนการเปลี่ยนแปลงสถานะของอ็อบเจ็กต์
- **Event** เป็นตัวกระตุ้นที่ทำให้เกิดการเปลี่ยนแปลงสถานะ เช่น การกดปุ่ม, การรับข้อมูล, การหมดเวลา
- **Guard/Condition** เป็นเงื่อนไขที่ต้องเป็นจริงเพื่อให้เกิดการเปลี่ยนแปลงสถานะ เช่น "ถ้าอุณหภูมิสูงกว่า 30 องศา"
- **Action**  การกระทำที่เกิดขึ้นเมื่อมีการเปลี่ยนแปลงสถานะ เช่น การแสดงข้อความ, การบันทึกข้อมูล

__ประเภทของ State Diagram__ 

- **Simple State Diagram**  แสดงสถานะและการเปลี่ยนแปลงสถานะแบบพื้นฐาน
- **Composite State Diagram**  แสดงสถานะที่ซับซ้อน โดยมีสถานะย่อย (Substate) อยู่ภายใน
- **State Machine Diagram**  เป็น State Diagram รูปแบบหนึ่งที่เน้นการทำงานแบบเครื่องจักรสถานะ (Finite State Machine)


__การใช้งาน State Diagram__

- อธิบายพฤติกรรมของอ็อบเจ็กต์ที่มีสถานะและการเปลี่ยนแปลงสถานะที่ซับซ้อน
- ออกแบบระบบควบคุม เช่น ระบบควบคุมเครื่องจักร, ระบบควบคุมการทำงานของโปรแกรม
- วิเคราะห์และแก้ไขปัญหาที่เกี่ยวข้องกับพฤติกรรมของระบบ

#### 4.1.4.5 Component Diagram
**Component Diagram** หรือ ไดอะแกรมส่วนประกอบ เป็นไดอะแกรมใน Unified Modeling Language (UML) ที่ใช้อธิบายโครงสร้างทางกายภาพของระบบซอฟต์แวร์ โดยแสดงให้เห็นส่วนประกอบ (Component) ต่างๆ และความสัมพันธ์ระหว่างส่วนประกอบเหล่านั้น ไดอะแกรมนี้ช่วยให้เห็นภาพรวมขององค์ประกอบซอฟต์แวร์ในระดับสูง และการพึ่งพาอาศัยกันของส่วนต่างๆ

**องค์ประกอบหลักของ Component Diagram**

- **Component**  แสดงด้วยสี่เหลี่ยมผืนผ้า มีสัญลักษณ์เป็นสี่เหลี่ยมเล็กๆ สองอันอยู่ด้านข้าง ส่วนประกอบแทนหน่วยซอฟต์แวร์ที่ทำงานได้ด้วยตัวเอง เช่น ไลบรารี, ไฟล์ปฏิบัติการ, ฐานข้อมูล ส่วนประกอบมีอินเทอร์เฟซ (Interface) สำหรับการสื่อสารกับส่วนประกอบอื่นๆ
- **Interface**  แสดงด้วยวงกลมเล็กๆ (เรียกว่า "Lollipop") หรือเส้นครึ่งวงกลม (เรียกว่า "Socket") อินเทอร์เฟซกำหนดชุดของการกระทำ (Operation) ที่ส่วนประกอบหนึ่งสามารถให้บริการหรือต้องการจากส่วนประกอบอื่น
    - *Provided Interface*   แสดงด้วยวงกลมทึบ (Lollipop) แสดงอินเทอร์เฟซที่ส่วนประกอบนั้นให้บริการแก่ส่วนประกอบอื่น
    - *Required Interface*   แสดงด้วยเส้นครึ่งวงกลม (Socket) แสดงอินเทอร์เฟซที่ส่วนประกอบนั้นต้องการจากส่วนประกอบอื่น
- **Port**  แสดงด้วยสี่เหลี่ยมเล็กๆ บนขอบของส่วนประกอบ เป็นจุดเชื่อมต่อระหว่างส่วนประกอบกับสภาพแวดล้อมภายนอก หรือกับส่วนประกอบอื่น พอร์ตสามารถเชื่อมต่อกับอินเทอร์เฟซ
- **Dependency**  แสดงด้วยเส้นประพร้อมหัวลูกศร แสดงว่าส่วนประกอบหนึ่งพึ่งพาการทำงานของอีกส่วนประกอบหนึ่ง

**ความสัมพันธ์ระหว่างส่วนประกอบ** 

- **Dependency** (การพึ่งพา)  ส่วนประกอบหนึ่งใช้อินเทอร์เฟซที่ให้บริการโดยอีกส่วนประกอบหนึ่ง
- **Assembly** (การประกอบ)  การเชื่อมต่อระหว่างอินเทอร์เฟซที่ต้องการของส่วนประกอบหนึ่งกับอินเทอร์เฟซที่ให้บริการของอีกส่วนประกอบหนึ่ง แสดงการทำงานร่วมกันของส่วนประกอบ

**การใช้งาน Component Diagram**

- อธิบายโครงสร้างทางกายภาพของระบบซอฟต์แวร์
- วางแผนการพัฒนาและการนำซอฟต์แวร์ไปใช้งาน
- แสดงการพึ่งพาอาศัยกันระหว่างส่วนประกอบต่างๆ
- สื่อสารการออกแบบซอฟต์แวร์กับผู้เกี่ยวข้อง

#### 4.1.4.6 Deployment Diagram

**Deployment Diagram** หรือ ไดอะแกรมการติดตั้ง เป็นไดอะแกรมใน Unified Modeling Language (UML) ที่ใช้อธิบายการกระจายหรือการติดตั้งซอฟต์แวร์บนฮาร์ดแวร์ โดยแสดงให้เห็นความสัมพันธ์ระหว่างส่วนประกอบซอฟต์แวร์ (Software Component) และโหนดฮาร์ดแวร์ (Hardware Node) ไดอะแกรมนี้ช่วยให้เห็นภาพรวมของสถาปัตยกรรมทางกายภาพของระบบ รวมถึงการพึ่งพาอาศัยกันของซอฟต์แวร์และฮาร์ดแวร์

**องค์ประกอบหลักของ Deployment Diagram**

- **Node**  แสดงด้วยกล่องสามมิติ แทนหน่วยประมวลผลทางกายภาพ เช่น เซิร์ฟเวอร์, คอมพิวเตอร์, อุปกรณ์เคลื่อนที่ หรืออุปกรณ์ฝังตัว โหนดอาจเป็น:
- **Device**  โหนดฮาร์ดแวร์ เช่น เซิร์ฟเวอร์, เดสก์ท็อป, โทรศัพท์มือถือ
- **Execution Environment**  โหนดซอฟต์แวร์ เช่น เว็บเซิร์ฟเวอร์, แอปพลิเคชันเซิร์ฟเวอร์, คอนเทนเนอร์
- **Component**  แสดงด้วยสัญลักษณ์เดียวกับ Component Diagram แทนส่วนประกอบซอฟต์แวร์ เช่น แอปพลิเคชัน, ไลบรารี, ฐานข้อมูล
- **Connection**  แสดงด้วยเส้นทึบระหว่างโหนด แสดงการสื่อสารหรือการเชื่อมต่อเครือข่ายระหว่างโหนด เช่น TCP/IP, HTTP
- **Artifact**  แสดงด้วยสี่เหลี่ยมผืนผ้าพร้อมคำว่า «artifact» อยู่ด้านบน แทนไฟล์หรือแพ็คเกจซอฟต์แวร์ที่ถูกติดตั้งบนโหนด เช่น ไฟล์ .exe, ไฟล์ .jar, ไฟล์ .war

**ความสัมพันธ์ระหว่างองค์ประกอบ**

- **Deployment** แสดงโดยการวางส่วนประกอบ (Component) หรือสิ่งประดิษฐ์ (Artifact) ไว้ภายในโหนด แสดงว่าส่วนประกอบนั้นถูกติดตั้งหรือใช้งานบนโหนดนั้น
- **Communication**  แสดงโดยการเชื่อมต่อระหว่างโหนด แสดงว่าโหนดเหล่านั้นสามารถสื่อสารกันได้

**การใช้งาน Deployment Diagram**

- อธิบายสถาปัตยกรรมทางกายภาพของระบบ
- วางแผนการติดตั้งและการกระจายซอฟต์แวร์
- แสดงการพึ่งพาอาศัยกันระหว่างซอฟต์แวร์และฮาร์ดแวร์
- สื่อสารการออกแบบระบบกับผู้เกี่ยวข้อง

#### 4.1.4.7 Timing Diagram
**Timing Diagram** หรือ แผนภาพเวลา เป็นไดอะแกรมใน Unified Modeling Language (UML) ที่ใช้อธิบายพฤติกรรมของระบบโดยเน้นที่การเปลี่ยนแปลงของสถานะหรือค่าของอ็อบเจ็กต์ต่างๆ ตามช่วงเวลา ไดอะแกรมนี้มีประโยชน์อย่างมากในการวิเคราะห์และออกแบบระบบแบบเรียลไทม์ (Real-time System) หรือระบบที่เกี่ยวข้องกับสัญญาณดิจิทัล

**องค์ประกอบหลักของ Timing Diagram**:
- **Lifeline** เส้นแนวนอนที่แสดงอ็อบเจ็กต์หรือสัญญาณที่สนใจ เช่น สัญญาณนาฬิกา, สัญญาณข้อมูล, สถานะของตัวแปร
- **State/Value**  แสดงบนเส้นชีวิต เป็นช่วงเวลาที่อ็อบเจ็กต์หรือสัญญาณมีสถานะหรือค่าใดๆ เช่น สูง (High), ต่ำ (Low), จริง (True), เท็จ (False), ค่าตัวเลข
- **Transition** แสดงด้วยเส้นแนวตั้งที่เชื่อมระหว่างสถานะหรือค่าที่แตกต่างกันบนเส้นชีวิต แสดงการเปลี่ยนแปลงสถานะหรือค่าของอ็อบเจ็กต์หรือสัญญาณ
- **Message** ลูกศรที่เชื่อมระหว่างเส้นชีวิต แสดงการส่งข้อความหรือสัญญาณระหว่างอ็อบเจ็กต์ต่างๆ
- **Timing Constraint**): แสดงด้วยสัญลักษณ์ต่างๆ เช่น ลูกศรสองหัวพร้อมตัวเลข แสดงช่วงเวลาที่เหตุการณ์ต้องเกิดขึ้น หรือเส้นหยักแสดงช่วงเวลาที่ไม่แน่นอน

**การใช้งาน Timing Diagram**

- อธิบายพฤติกรรมของระบบแบบเรียลไทม์
- วิเคราะห์การทำงานของวงจรอิเล็กทรอนิกส์หรือระบบดิจิทัล
- ตรวจสอบข้อกำหนดด้านเวลาของระบบ
- แสดงลำดับเหตุการณ์และการเปลี่ยนแปลงสถานะตามเวลา

#### 4.1.4.8 Network diagram
**Network Diagram** โดยทั่วไปแล้วไม่ได้เป็นส่วนหนึ่งของ UML (Unified Modeling Language) โดยตรง UML เน้นการสร้างแบบจำลองซอฟต์แวร์ ไม่ใช่โครงสร้างเครือข่ายทางกายภาพ อย่างไรก็ตาม เราสามารถใช้ไดอะแกรมบางประเภทใน UML เพื่อแสดงภาพบางแง่มุมของเครือข่าย โดยเฉพาะอย่างยิ่งในบริบทของการนำซอฟต์แวร์ไปใช้งาน

ไดอะแกรมที่มักใช้เพื่อแสดงภาพเครือข่ายในบริบทของ UML คือ Deployment Diagram  (รายละเอียดอยู่ในหัวข้อ 4.1.4.6)
